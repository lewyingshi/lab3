```python
# square root function
sq_root = lambda x: x**2
sq_root(8)
```

```python
# edge case
sum_of_squares([1])
```

```python
# lambda function
sum_of_squares = lambda num: sum([sq_root(num) for num in num[:]])
sum_of_squares([2,3,4])
```

```python
sum_of_squares(range(5))
```

```python
sum_of_squares([3,3,3])
```

```python
# helper function
sq_root = lambda x: x**2

# test function
def sum_of_sq(myList):
    """
    """
    output = sum([sq_root(myList) for myList in myList[:]])
    return output

def test_sum_of_sq():
    assert sum_of_sq([1]) == 1
    assert sum_of_sq(range(5)) == 30
    assert sum_of_sq([3, 3, 3])
test_sum_of_sq()
```
